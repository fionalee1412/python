### Series

```python
# 数组中的数据可以是任意的类型，包括整数、浮点数、字符串、列表、字典等
# 数据应为同一类型
a = [1,2,2,3]
b = [3,4,5,6]
s = pd.Series([a,b])
s
dict1 = {"b":[1,2,3],"a":[2,3,4]}
dict1
s1 = pd.Series(dict1)
s1
```

```python
# Series的属性
s1.index = ['li','fang']
s1
# 不允许修改单个index，只能一块改了
s1.values
```

```python
# 索引index
s2 = pd.Series(np.arange(10),index=list('abcdefghij'))
s2[-1]
s2[0]
s2[[0,1,3]]
s2["a"]
```

```python
# 切片Slice
s2[0:3] # 没有包含末端
s2["a":"f"] # 包含了末端
```

```python
# 修改Series的值
s2[2] = 12
s2["a"] = 1
s2
```

```python
# 添加值
s2.append(pd.Series([50,60],index=['a1','a2']))
s2
# append没有修改原来的series
s2["a1"] = [0,1]
s2 # 只能一次添加一个
```

```python
# 删除值
#del s2[1]
s2
#del s2[0]
```

```python
# 过滤
s2[s2 > 5]
s2[s2%2 ==0]
s2[s2%2 !=0]
```

```python
# 缺失值处理
s3 = s2.append(pd.Series([np.nan,None],index=["h",'k']))
s3
s3.isnull()
s3[s3.isnull()]
s3.dropna() 
s3 # dropna会返回一个新的Series
s3[~s3.isnull()] # ~表示否定，即返回非null值
s3.fillna(value=2) 
s3 # 还是会返回一个新的Series
s3.fillna(value=2,inplace=True) # 会替换原来的Series
s3
s = pd.Series([5,np.nan,10,30,None,None])
s
s.fillna(method='ffill') # 意思是，null值取它前面的那个值
s.fillna(method='bfill') #这个是null值取排在它后面的那个值
```

```python
# 排序
s2.sort_index() # 按照索引升序
s2.sort_values(ascending=False) # 按照值降序
```

```python
# 描述性哦统计
s5 = pd.Series([100,100,2,40,40,40,30,50])
s5.value_counts() # 值的计数
s5.min() # 最小值
s5.max() #最大值
s5.median() # 中位数
s5.mean() #均值
s5.sum()
s5.std()
s5.describe().round(1)
```

```python
# 向量化运算
s5+1000
s5 # 不会修改原Series
s5*2
```
