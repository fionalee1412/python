```python
from urllib import urlopen

# 用decode显示出正常中文的额样式
html = urlopen('http://www.boyaa.com/page/news/newsinfo68.html').read().decode('utf-8')
print(html)
```

```python
# 用正则匹配
import re
# re.findall(pattern,string,flags=0) 返回一个列表
res = re.findall(r"<title>(.*)</title>",html)
print(res[0])
```

```python
# 用flags=re.DOTALL对tab、new line不敏感
res = re.findall(r"<p>(.*)</p>",html,flags=re.DOTALL)
print(res[0])
```

```python
# 注意匹配双引号要外面要用单引号
res = re.findall(r'href="http://(.*?)"',html)
res
```

## 爬取百度百科

```python
import random
base_url = 'https://baike.baidu.com'
his = ["/item/大王具足虫#hotspotmining"]
for i in range(10):
    url = base_url+his[-1]
    html = urlopen(url).read().decode("utf-8")
    baike = BeautifulSoup(html,features="lxml")
    print(baike.find('h1').get_text())
    print(base_url+his[-1])

    # 选取需要的url连接
    sub_urls = baike.find_all("a",{"target":"_blank","href":re.compile("^/item/(%.{2})+$")})

    # 从这些页面中随机选取一个，当做下一个要爬的页面
    if len(sub_urls) !=0:
        his.append(random.sample(sub_urls,1)[0]['href'])
    else:
        his.pop()
```

## 练习：下载图片

```python
url = 'https://www.douban.com/note/673422259/'
# 用requests获取链接
html = requests.get(url).text
soup = BeautifulSoup(html,'lxml')
img_url = soup.find_all('div',class_="photo_wrap")
#page = soup.find('div',class_="paginator").find_all("a",href=re.compile("https://img3.doubanio.com/view/note/l/public/"))


for ul in img_url:
    imgs = ul.find_all('img')
    for img in imgs:
        url = img['src']
        r = requests.get(url,stream=True)
        image_name = url.split('/')[-1]
        with open('./img/%s'%image_name,'wb')as f:
            for chunk in r.iter_content(chunk_size=128):
                f.write(chunk)
            print('%s'%image_name)
```
