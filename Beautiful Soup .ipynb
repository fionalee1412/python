## 快速开始
        

```python
from urllib import urlopen
from bs4 import BeautifulSoup
html = urlopen("https://movie.douban.com/cinema/nowplaying/shenzhen/").read().decode("utf-8")
soup = BeautifulSoup(html,"lxml")
print(soup.prettify())
```

```python
soup.title
soup.title.name
soup.title.string
soup.title.parent.name
soup.p
#soup.find_all('a')
soup.find_all(id='doubanapp-tip')
```

```python
for link in soup.find_all('img'):
    print(link.get('src'))
    print(link.get('alt'))
```

## find_all

```python
# 过滤器
soup.find_all("h1")
# 正则表达式
for tag in soup.find_all(re.compile("^t")):
    print(tag.text)
    
# 列表，会与任一元素匹配的内容返回
soup.find_all(['h1','h2'])
```

#### find_all(name,attrs,recursive,text,**kwargs)
   搜索当前tag的所有tag子节点

```python
soup.find_all("title")
# 相当于li class='stitle'
#soup.find_all("li","stitle")
a = soup.find_all("span","subject-rate")
```

#### name参数
可以查找所有名字为name的tag

#### keyword参数
tag里面的属性

```python
#soup.find_all(id='content')
soup.find_all(href=re.compile("playing_poster$"))
```

```python
soup.find_all("a",limit=2)
```

### recursive参数
设置为false的时候，会只检索tag的直接子节点

```python
soup.find_all('li',recursive=False)
```
