# 定义问题
> 异常用户包括：
* 刷号用户（获取系统的免费游戏币）
* 广告用户（在游戏内通过头像、发言等发布广告，不玩牌，严重影响体验）
* 作弊用户（一般是多个账号一起，通过通牌作弊）  

* 这些用户严重影响了正常玩家的玩牌体验，同时导致了公司受损，一定要去除掉
* 不同的作弊类型对应不同的行为特征，之前通过人工导数据来判定，效率低下，而且由于游戏准许游客登录，注册新号的成本非常低，导致这样的用户非常多 
* 刷号和广告基本上都是通过脚本控制，行为较为机械，有固定的模型，相对好识别  
* 作弊用户往往不容易被发现，往往是通过用户投诉，然后运营/客服通过查询其行为，判定是否有作弊。这里可以通过机器学习代替人工审核的过程，因为判定的过程也是有一定的流程的；并且可以通过之前留下的数据进行训练。

## 预测变量
* #### 广告用户和刷号用户  
这是一个典型的二分类问题，即是否判定为广告/刷号
* #### 作弊用户  
这里即可以当做二分类，也可以是回归，但是回归会比二分类要更贴合业务，因为为了避免误封号造成影响，还是需要人工判定，我们可以给出一个作弊分值，在分值高于某个值，可以直接判定为作弊，如果分值不高，可以通过人工二次审核来标记
